```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
picture=io.imread('smile.jpg')
picture.shape

```

```python
plt.imshow(picture)
```

#Edit Image

#Horizontal Flip

```python
from numpy import fliplr, flipud
vertical_flip = flipud(picture)
plt.imshow(vertical_flip)
```

```python
from numpy import fliplr, flipud
vertical_flip = fliplr(picture)
plt.imshow(vertical_flip)
```

#Invert Colors

```python
from PIL import Image
import numpy as np
img = Image.open(r"smile.jpg")
img_arry = np.array(img)
I_max = 255
img_arry = I_max - img_arry
inverted_img = Image.fromarray(img_arry)
inverted_img.save(r"simle_inv.jpg")
```

#convert to grayscale

```python
import skimage
gray_image = skimage.color.rgb2gray(picture)
fig, ax = plt.subplots()
plt.imshow(gray_image, cmap="gray")
```

#Convert to B&W

```python
from PIL import Image
img=Image.open('smile.jpg')
cn=img.convert("L")
cn.save("smile1.jpg")
cn.show()
```
